```python
import pandas as pd
import glob
```

```python
datasets = glob.glob("./songs_wikipedia_final/*.csv")
datasets2 = glob.glob("./songs_lyrics_final/*.csv")
```

```python
count=0
for i in datasets2:
    state = i.split("/")[-1].strip()
    file_loc = "./songs_wikipedia_final/ "+state

    if file_loc in datasets:
        count+=1
        df1 = pd.read_csv(i)
        df2 = pd.read_csv(file_loc)
        
        new_df = pd.concat([df1,df2],axis =0)
    else:
        new_df = pd.read_csv(i)
    new_df = new_df[~new_df['danceability'].isnull()]
    new_df = new_df.loc[:, ~new_df.columns.str.contains('^Unnamed')]
    new_df = new_df.loc[:, ~new_df.columns.str.contains('^error')]
        
    new_df.to_csv("./merged_final/"+state)
```
