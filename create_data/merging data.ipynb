```python
import pandas as pd
import glob
```

```python
datasets = glob.glob("./songs_wikipedia_final/*.csv")
datasets2 = glob.glob("./songs_lyrics_final/*.csv")
```

```python
for i in datasets:
    state = i.split("/")[-1].strip()
    file_loc = "./songs_lyrics_final/"+state
    if file_loc in datasets2:
        df1 = pd.read_csv(i)
        df2 = pd.read_csv(file_loc)
        
        new_df = pd.concat([df1,df2],axis =0)
        print(new_df)
        break
```
